```python
# ASSIGNMENT 2
n=int(input("enter the no of tuples u want to enter"))
lis=[]
for i in range(n):
  x=eval(input("enter two values of tuple seperated by coma"))
  lis.append(x)
print("the list of tuple given as input is",lis)  
d=dict(lis)
li=[]
for x,y in d.items():
  s=y+x
  li.append(s)
print("the final output is ",li)  
```
